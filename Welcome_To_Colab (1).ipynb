```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import xgboost as xgb
```

```
 data = pd.read_csv('house_price01.csv')
```

```
display(data)
```

```
data.shape
```

```
data.info()
```

```
data.describe()
```

```
data = data.fillna(method='ffill')
data
```

```
x = data.drop(columns=['Price'])
y = data['Price']
```

```
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=2)
print('80% input data size is', xtrain.shape)
print('20% input data size is', xtest.shape)
print('80% output data size is', ytrain.shape)
print('20% output data size is', ytest.shape)
```

```
x.shape
```

```
model = xgb.XGBRegressor(objective='reg:squarederror',n_estimators=100)
model.fit(xtrain, ytrain)
```

```
output = model.predict(xtest)
display(output)
```

```
abs(r2_score(ytest, output))
```
